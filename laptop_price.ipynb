```python
import requests
from bs4 import BeautifulSoup
import httplib2
import webbrowser 
from pprint import pprint
```

```python
# Define the URL of the website you want to scrape
url = "https://www.smartprix.com/laptops"
```

```python
# Send an HTTP GET request to the URL
response = requests.get(url)
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find laptop prices using appropriate HTML tags and attributes
    laptop_prices = soup.find_all('div', class_='laptop-price')  # Update with the actual HTML structure

    # Process and print the laptop prices
    for laptop_price in laptop_prices:
        price = laptop_price.text.strip()  # Extract and clean the price text
        print(f"Laptop Price: {price}")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```
